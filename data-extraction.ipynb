### Cleaning of data

```python
#Extracting the variables with type Object from the df
df_object = pd.DataFrame(df[['Gender', 'family_history_with_overweight','FAVC', 'CAEC', 'SMOKE', 'SCC', 'CALC', 'MTRANS'
                         ,'NObeyesdad']])
df_object.head()
```

```python
#Convert the variable types from object to category
df_cat = df_object.astype("category")

df_cat.describe()
```

```python
df.head()
```

```python
#Extracting the numeric variables from the original df
df_numeric = df.drop(labels = ['Gender', 'family_history_with_overweight','FAVC', 'CAEC', 'SMOKE', 'SCC', 'CALC', 'MTRANS'
                         ,'NObeyesdad'], axis = 1)
```

```python
print(df_numeric.dtypes)
print(df_numeric.head())
```

```python
#Joining both numeric and categorical variables into one dataframe called 'df_catData'
#Basically the same as the original df, just that object is converted to categorical
df_catData = df_numeric.join(df_cat)
df_catData.head()
```

```python
print(df_catData.dtypes)
```

```python
count_nan_in_df = df_catData.isnull().sum().sum()
print("Count of Nan: " + str(count_nan_in_df))
```

### Export all df to csv 
this is to ensure that the df can be accessed in other notebooks

```python
#reusable function to save df to csv so that the df can be accessable to all notebooks
def save_csv(df, folder_path ,file_name ):
    file_path = os.path.join(folder_path, file_name)
    df.to_csv(file_path, index=False)
    print(f"Successfully saved df {file_name} to {folder_path}")
```

```python
#uploaded of df to csv
#catData = original dataset [the change from dtype -> object to cat]
#obesityVars = df with only the needed attributes
save_csv(df_catData,'datasets', 'original_eating_habits_data.csv')
save_csv(df_obesityVars,'datasets', 'cleaned_eating_habits_data.csv')
save_csv(df_obesityVars_cat,'datasets', 'cat_eating_habits_data.csv')
save_csv(df_obesityVars_int,'datasets', 'int64_eating_habits_data.csv')
```
